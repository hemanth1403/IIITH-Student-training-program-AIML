## 1. Convolutional Operations

Central to Convolutional Neural Networks (CNN), a convolution operation is a linear operation which involves element-wise multiplication between a small filter (say, a matrix of integers) and filter-sized patch from the image. We move this filter across the image like a sliding window from top left to bottom right. For each point on the image, a value is calculated based on the filter using a convolution operation. These filters can do simplest task like checking if there is a vertical line in the image or complicated task like detecting a human eye in the image.

Let's look at the convolution formula:

Convolution between image $f(x, y)$ and kernel $k(x, y)$ is
$$f(x,y) * k(x,y) = \sum \limits _{i=0} ^{W-1} \sum \limits _{j=0} ^{H-1} f(i, j) k(x − i, y − j)$$

where $W$ and $H$ are the the width and height of the image.

```python
# Vertical Line filter
filter = np.array([[1,0,-1],
                   [1,0,-1],
                   [1,0,-1]])
```

Non-zero output suggests that there is a vertical edge present in the first image and not present in the second image.

As expected, the horizontal edge is detected in second image with this filter.

Now, we will take a bigger image (5 x 5) and see how a convolution operation works by sliding a filter left to right and top to bottom to obtain an output map from image. Let's define a function ***apply_filter()*** for this.

```python
def apply_filter(img, filter):
  height, width = img.shape
  filter_size = filter.shape

  output = np.empty(0)

  # Move the filter over entire image and store the result in output
  for i in range(0, height - filter_size[1] + 1):
    for j in range(0, width - filter_size[0] + 1):
      # Matrix multiplication for a single patch of image and filter
      output = np.append(output, np.sum(np.multiply(img[i:i+filter_size[0], j:j+filter_size[1]], filter)))

  # Calculate the output shape of the resultant image
  output_shape = (height - (filter_size[1]-1)), (width - (filter_size[0]-1))

  # Return the reshaped image
  return output.reshape(output_shape)
```

Plotting function

```python
def plot_images(images, titles, tick_params=True):
  n = len(images)
  fig = plt.figure(figsize=(10,4))
  for i in range(n):
    ax = fig.add_subplot(1,n,i+1)
    if len(images[i].shape) == 2:
      ax.imshow(images[i], cmap='gray',
                extent=(0,images[i].shape[1], images[i].shape[0], 0))
    else:
      ax.imshow(images[i])
    ax.set_title(titles[i])
    if not tick_params:
      plt.tick_params(axis='both', labelbottom=False, bottom=False,
                labelleft=False, left=False)
  plt.show()
```

```python
# 2D image
img = np.array([[20,20,0,0,0],
                [20,20,0,0,0],
                [20,20,0,0,0],
                [20,20,0,0,0],
                [20,20,0,0,0]])

# Vertical edge filter
filter = np.array([[1,0,-1],
                   [1,0,-1],
                   [1,0,-1]])


output = apply_filter(img, filter)
print(output) # Note the shape of output image!
```

```python
# Let's plot the above image with results
images = []
titles = []

images.append(img)
titles.append('Original Image')

images.append(filter)
titles.append('Filter')

images.append(output)
titles.append('Convolution Output')

plot_images(images, titles)
```

As, you can see, horizontal edge is detected in the output.

```python
# 1. A

def apply_filter_with_padding(img, filter, padding_type='constant', padding_value=0):
    height, width = img.shape
    filter_size = filter.shape

    # Calculate the padding sizes
    pad_height = filter_size[1] - 1
    pad_width = filter_size[0] - 1

    # Apply padding to the image
    img_padded = np.pad(img, ((0, pad_height), (0, pad_width)), mode=padding_type, constant_values=padding_value)

    output = np.empty(0)

    # Move the filter over the entire image and store the result in output
    for i in range(0, height):
        for j in range(0, width):
            # Matrix multiplication for a single patch of image and filter
            output = np.append(output, np.sum(np.multiply(img_padded[i:i+filter_size[0], j:j+filter_size[1]], filter)))

    # Calculate the output shape of the resultant image
    output_shape = (height, width)

    # Return the reshaped image
    return output.reshape(output_shape)

image = cv2.imread('lotus.jpg', 0)
print('Original image size: ', image.shape)

# Saving images for plots
images = [image]
titles = ['Original Image']

# Vertical edge filter
filter = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])

images.append(filter)
titles.append('Filter')

# Apply this filter to the image with padding
output = apply_filter_with_padding(image, filter, padding_type='constant', padding_value=0)

print('Output image size with padding: ', output.shape)

images.append(output)
titles.append('Convolution Output with Padding')

plot_images(images, titles)
```

```python
# 2. A

def apply_filter_with_padding_and_stride(img, filter, padding_type='constant', padding_value=0, stride=1):
    height, width = img.shape
    filter_size = filter.shape

    # Calculate the padding sizes
    pad_height = filter_size[1] - 1
    pad_width = filter_size[0] - 1

    # Apply padding to the image
    img_padded = np.pad(img, ((0, pad_height), (0, pad_width)), mode=padding_type, constant_values=padding_value)

    output = np.empty(0)

    # Move the filter over the entire image with stride and store the result in output
    for i in range(0, height, stride):
        for j in range(0, width, stride):
            # Matrix multiplication for a single patch of image and filter
            output = np.append(output, np.sum(np.multiply(img_padded[i:i+filter_size[0], j:j+filter_size[1]], filter)))

    # Calculate the output shape of the resultant image
    output_shape = ((height - 1) // stride + 1, (width - 1) // stride + 1)

    # Return the reshaped image
    return output.reshape(output_shape)


# Read the image with OpenCV
image = cv2.imread('lotus.jpg', 0)
print('Original image size: ', image.shape)

# Saving images for plots
images = [image]
titles = ['Original Image']

# Vertical edge filter
filter = np.array([[1, 0, -1],
                   [1, 0, -1],
                   [1, 0, -1]])

images.append(filter)
titles.append('Filter')

# Apply this filter to the image with padding and stride
output = apply_filter_with_padding_and_stride(image, filter, padding_type='constant', padding_value=0, stride=2)

print('Output image size with padding and stride: ', output.shape)

images.append(output)
titles.append('Convolution Output with Padding and Stride')

# Let's plot the images
plot_images(images, titles)
```

Now lets implement a CNN in pytorch and use the learnt representations for image classification of MNIST dataset.

```python
# Import packages
import torch
import torchvision
import torch.nn as nn
from torch import optim
import torch.nn.functional as F

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Device configuration (whether to run on GPU or CPU)
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
# Set seeds for reproducibility
seed = 0
np.random.seed(seed)
torch.manual_seed(seed)
torch.cuda.manual_seed(seed)
torch.backends.cudnn.deterministic = True
```

#### Load MNIST data
We will use the [MNIST dataset](https://pytorch.org/vision/stable/datasets.html#mnist) from torchvision Pytorch and setup the train and test dataloaders.

```python
batch_size_train = 128
batch_size_test = 128

# Images in torchvision datasets are PIL Images in range [0,1] so we need
# 'ToTensor' transform to convert them into tensors
train_data = torchvision.datasets.MNIST('./data', train=True, download=True,
                             transform=torchvision.transforms.ToTensor())
test_data = torchvision.datasets.MNIST('./data', train=False, download=True,
                             transform=torchvision.transforms.ToTensor())

train_loader = torch.utils.data.DataLoader(train_data, batch_size=batch_size_train, shuffle=True)
test_loader = torch.utils.data.DataLoader(test_data, batch_size=batch_size_test, shuffle=False)
```

#### Understand the dataset
Let us now visualize the dataset in terms of number of samples, classes etc.

```python
print('Training data shape : ', train_data.data.shape, train_data.targets.shape)
print('Testing data shape : ', test_data.data.shape, test_data.targets.shape)

# Find the unique numbers from the train labels
classes = np.unique(train_data.targets.numpy())
nClasses = len(classes)
print('Total number of outputs : ', nClasses)
print('Output classes : ', classes)
```

```python
# Helper function to plot data
def plot_data(images, labels, classes=None):
  figure = plt.figure(figsize=(9, 4))
  cols, rows = 5, 2
  for i in range(1, cols * rows + 1):
      sample_idx = torch.randint(len(images), size=(1,)).item()
      img, label = images[sample_idx], labels[sample_idx]
      figure.add_subplot(rows, cols, i)
      if classes is not None:
        label = classes[label]
      plt.title('Label:' +str(label))
      plt.axis("off")
      plt.imshow(img.squeeze(), cmap="gray")
  plt.show()
```

```python
plot_data(train_data.data, train_data.targets.numpy())
```

#### Define the model

```python
class CNN(nn.Module):
    def __init__(self):
        super(CNN, self).__init__()
        self.conv1 = nn.Conv2d(in_channels=1, out_channels=32, kernel_size=3, stride=1, padding=1)
        self.max_pool1 = nn.MaxPool2d(kernel_size=2, stride=2)
        self.conv2 = nn.Conv2d(32, 64, 3, 1, 1)
        self.max_pool2 = nn.MaxPool2d(kernel_size=2, stride=2)

        # fully connected layer
        self.fc = nn.Linear(64 * 7 * 7, 128)
        # output layer 10 classes
        self.out = nn.Linear(128, 10)

    def forward(self, x):
        x = self.conv1(x)
        x = F.relu(x) #activation
        x = self.max_pool1(x)
        x = self.conv2(x)
        x = F.relu(x)
        x = self.max_pool2(x)
        # flatten the output for FC layer
        x = x.view(x.size(0), -1)
        x = self.fc(x)
        output = self.out(x)
        return output
```

```python
# Build the model object and put on the device
model = CNN().to(device)
print(model)
```

#### Define Loss function

```python
# Cross Entropy loss for multi-class classification
loss_func = nn.CrossEntropyLoss()
```

#### Define optimizer

```python
# Basic SGD optimizer with 0.01 learning rate
optimizer = optim.SGD(model.parameters(), lr = 0.01)
```

#### Train the model

Helper function for training/testing

```python
def train(num_epochs, model, train_loader, loss_func, optimizer):

  # Training mode
  model.train()

  train_losses = []
  train_acc = []

  # Train the model
  for epoch in range(num_epochs):
    running_loss = 0
    running_acc = 0
    for i, (images, labels) in enumerate(train_loader):

      # clear gradients for this training step
      optimizer.zero_grad()

      # Put data on devices
      images = images.to(device)
      labels = labels.to(device)

      # Forward pass
      output = model(images)

      # Calculate loss
      loss = loss_func(output, labels)

      # Backpropagation, compute gradients
      loss.backward()

      # Apply gradients
      optimizer.step()

      # Running loss
      running_loss += loss.item()

      # indices of max probabilities
      _, preds = torch.max(output, dim=1)

      # Calculate number of correct predictions
      correct = (preds.float() == labels).sum()
      running_acc += correct

      epoch_loss = running_loss / len(train_loader.dataset)
      epoch_acc = running_acc / len(train_loader.dataset)

    train_losses.append(epoch_loss)
    train_acc.append(epoch_acc)
    print ('Epoch {}/{}, Loss: {:.4f}, Accuracy: {:.4f}'.format(epoch + 1, num_epochs, epoch_loss, epoch_acc*100))

  return train_losses, train_acc
```

```python
def test(model, test_loader):
  # Eval mode
  model.eval()
  test_acc = 0
  correct = 0
  for i, (images, labels) in enumerate(test_loader):
    # Deactivate autograd engine (don't compute grads since we're not training)
    with torch.no_grad():
      images, labels = images.to(device), labels.to(device)
      output = model(images)

      # Calculate number of correct predictions
      _, preds = torch.max(output, dim=1)
      correct += (preds == labels).sum()

  test_acc = correct / len(test_loader.dataset)
  print('Test Accuracy: {:.4f}'.format(test_acc*100))

  # Plot the images with predicted labels
  plot_data(images.data.cpu().numpy(), preds.data.cpu().numpy(), test_loader.dataset.classes)
```

Start training

```python
num_epochs = 10  # iterations
train_losses, train_acc = train(num_epochs, model, train_loader, loss_func, optimizer)
```

Plot training plots

```python
fig = plt.figure(figsize=(10,4))
ax = fig.add_subplot(1,2, 1)
ax.plot(np.arange(1,len(train_losses)+1),train_losses)
plt.xlabel('Training loss')
plt.ylabel('Epochs')
ax.set_title('Loss vs Epochs')
plt.show()
```

#### Evaluate the model

```python
# Evaluate the model on testing data and plot predictions
test(model, test_loader)
```

Q 1: What is the ratio of parameters in single 5 x 5 kernel and equivalent stacked 3 x 3 kernels? Consider number of channels in input and output channels as C.

## Ans.
The number of parameters in a convolutional layer depends on the size of the kernel, the number of input channels, and the number of output channels.

For a single 5 x 5 kernel with C input and C output channels, the number of parameters would be:

$$\text{Parameters}_{5x5} = 5 * 5 * C * C$$

For two stacked 3 x 3 kernels with C input and C output channels, the number of parameters would be:

$$\text{Parameters}_{3x3} = 2 * (3 * 3 * C * C)$$

Therefore, the ratio of parameters in a single 5 x 5 kernel to the equivalent stacked 3 x 3 kernels is:

$$\text{Ratio} = \frac{\text{Parameters}_{5x5}}{\text{Parameters}_{3x3}} = \frac{5 * 5}{2 * 3 * 3} = \frac{25}{18}$$

So, the ratio is 25:18. This means that using two stacked 3 x 3 kernels instead of a single 5 x 5 kernel can reduce the number of parameters, thereby reducing the computational complexity. This is one of the reasons why modern convolutional neural networks often use multiple stacked smaller kernels instead of a single larger kernel.


Q 2: How can you replace 7 x 7 convolution kernel using only 3 x 3 kernels? What would be ratio of parameters in this case? Consider number of channels in input and output channels as C.

## Ans.

A 7 x 7 convolution kernel can be replaced by three consecutive 3 x 3 convolution kernels. This is because the receptive field of three 3 x 3 convolutions is the same as that of one 7 x 7 convolution.

For a single 7 x 7 kernel with C input and C output channels, the number of parameters would be:

$$\text{Parameters}_{7x7} = 7 * 7 * C * C$$

For three stacked 3 x 3 kernels with C input and C output channels, the number of parameters would be:

$$\text{Parameters}_{3x3} = 3 * (3 * 3 * C * C)$$

Therefore, the ratio of parameters in a single 7 x 7 kernel to the equivalent stacked 3 x 3 kernels is:

$$\text{Ratio} = \frac{\text{Parameters}_{7x7}}{\text{Parameters}_{3x3}} = \frac{7 * 7}{3 * 3 * 3} = \frac{49}{27}$$

So, the ratio is 49:27. This means that using three stacked 3 x 3 kernels instead of a single 7 x 7 kernel can reduce the number of parameters, thereby reducing the computational complexity. This is one of the reasons why modern convolutional neural networks often use multiple stacked smaller kernels instead of a single larger kernel.


## 3. CNN Visualization

Save the conv layers and their weights

```python
model_weights = [] # we will save the conv layer weights in this list
conv_layers = [] # we will save the conv layers in this list
# get all the model children as list
model_children = list(model.children())

# counter to keep count of the conv layers
counter = 0
# append all the conv layers and their respective weights to the list
for i in range(len(model_children)):
    if type(model_children[i]) == nn.Conv2d:
        counter += 1
        model_weights.append(model_children[i].weight)
        conv_layers.append(model_children[i])
    elif type(model_children[i]) == nn.Sequential:
        for j in range(len(model_children[i])):
            for child in model_children[i][j].children():
                if type(child) == nn.Conv2d:
                    counter += 1
                    model_weights.append(child.weight)
                    conv_layers.append(child)
print(f"Total convolutional layers: {counter}")
```

```python
# take a look at the conv layers and the respective weights
for weight, conv in zip(model_weights, conv_layers):
    # print(f"WEIGHT: {weight} \nSHAPE: {weight.shape}")
    print(f"CONV: {conv} ====> SHAPE: {weight.shape}")
```

### Exercises

Q: List a few practical applications of convolutional autoencoders.

## Ans.

1. **Image Compression**: Convolutional autoencoders can compress high-dimensional image data into a lower-dimensional data. This improves storage efficiency and transmission of image data.

2. **Image Reconstruction**: They can reconstruct missing parts of an image. It can also handle images with slight variations in object position or orientation.

3. **Anomaly Detection**: Autoencoders can accurately capture the key aspects of the provided data to flag anomalies.

4. **Data Visualization**: Autoencoders compress a high-dimensional dataset into a lower-dimensional representation that can be utilized for data visualization.

5. **Pattern Recognition**: They are used in fields where the specific object and its precise boundary detection are essential.

6. **Dimensionality Reduction**: Autoencoders are typically used for dimensionality reduction, feature learning, and data compression.



Q: What change do we need to make for the autoencoder to reduce into PCA?

## Ans.

1. **Linear Activation Function**: Change all the activation functions in the autoencoder to linear. This is because PCA is a linear transformation, while autoencoders can have nonlinear encoder/decoders.

2. **Single Layer**: Use a single layer autoencoder. A single layer autoencoder with a linear transfer function is nearly equivalent to PCA.

3. **Bottleneck Architecture**: Form a bottleneck in the network. This is done by using a neural network with a narrow hidden layer.

Even after these changes, the weight matrix found by the autoencoder and PCA won't necessarily be the same, but the subspace spanned by the respective weight matrices will be. This is because PCA identifies a matrix that achieves the best possible square error.
